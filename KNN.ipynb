```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,accuracy_score
import numpy as np
```

```python
df=pd.read_excel('train_data.xlsx')
df1=pd.read_excel('test_data.xlsx')
```

```python
train1=df['expected_glucose']
test1=df1['expected_glucose']
```

```python
train=df.drop('expected_glucose',axis=1)
test=df1.drop('expected_glucose',axis=1)
```

```python
from sklearn.pipeline import Pipeline

error=[]
for i in range(1,51):
    num_neigh = i
    neigh = KNeighborsRegressor(n_neighbors=num_neigh)
    neigh.fit(train ,train1 )
    predict_train = neigh.predict(train)
    mse_on_train = np.sqrt(mean_squared_error(train1 , predict_train))
    print("R.M.S.E on training data"+" at k=",num_neigh,"is" , mse_on_train)
    predict_test = neigh.predict(test)

    mse_on_test = mean_squared_error(test1 , predict_test)
    error.append(np.sqrt(mse_on_test))
```

```python
for i in range(1,51):
    print("R.M.S.E on testing data at k =",i,"is",error[i-1])
```

```python
import matplotlib.pyplot as plt
x = [i for i in range(1,51)]
plt.plot(x , error)
plt.show()
print(min(error))
```
